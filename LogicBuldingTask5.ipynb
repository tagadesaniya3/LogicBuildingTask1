```python
def analyze_post_engagement(likes_list):
    total_likes = 0

    # Loop to calculate total likes
    for likes in likes_list:
        total_likes += likes

    print("Total Likes:", total_likes)

    # Check if post is viral
    if total_likes >= 1000:
        print("Post Status: Viral Post")
    else:
        print("Post Status: Normal Engagement")


# Example usage
likes_data = [200, 300, 150, 400]  # Sample likes from different sources
analyze_post_engagement(likes_data)
```

```python
def check_rainfall(rainfall_data, required_level):
    total_rainfall = 0

    # Loop to calculate total rainfall
    for rain in rainfall_data:
        total_rainfall += rain

    # Calculate average rainfall
    average_rainfall = total_rainfall / len(rainfall_data)

    print("Average Rainfall:", int(average_rainfall))

    # Check rainfall adequacy
    if average_rainfall >= required_level:
        print("Rainfall Status: Adequate Rainfall")
    else:
        print("Rainfall Status: Inadequate Rainfall")


# Example usage
seasonal_rainfall = [70, 75, 68, 75]  # Sample rainfall data
required_rainfall_level = 70

check_rainfall(seasonal_rainfall, required_rainfall_level)
```

```python
def detect_duplicate_accounts(usernames):
    unique_usernames = set(usernames)

    if len(unique_usernames) < len(usernames):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")


# Example usage
accounts = ["john123", "emma_w", "john123", "alex99"]
detect_duplicate_accounts(accounts)
```

```python
def filter_premium_crops(crop_prices):
    premium_crops = []

    # Loop to filter premium crop prices
    for price in crop_prices:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Example usage
prices = [1500, 2500, 1800, 3200]
filter_premium_crops(prices)
```

```python
def check_attendance_eligibility(attendance_record):
    total_classes = len(attendance_record)
    attended_classes = 0

    # Loop to count attended classes
    for attended in attendance_record:
        if attended:  # True means present, False means absent
            attended_classes += 1

    # Calculate attendance percentage
    attendance_percentage = (attended_classes / total_classes) * 100
    print("Attendance Percentage:", attendance_percentage)

    # Check exam eligibility
    if attendance_percentage >= 75:
        print("Exam Eligibility: Eligible")
    else:
        print("Exam Eligibility: Not Eligible")


# Example usage
attendance = [True, True, False, True, True, True, True, False, True, True]  # Sample attendance record
check_attendance_eligibility(attendance)
```

```python
def calculate_electricity_bill(units):
    # Initialize bill
    bill = 0

    # Calculate bill based on slabs
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        # First 100 units @3, remaining @5
        bill = 100 * 3 + (units - 100) * 5
    else:
        # First 100 units @3, next 100 units @5, remaining @7
        bill = 100 * 3 + 100 * 5 + (units - 200) * 7

    # Determine usage classification
    if bill < 500:
        usage_status = "Low Usage"
    elif 500 <= bill <= 1500:
        usage_status = "Moderate Usage"
    else:
        usage_status = "High Usage"

    return bill, usage_status


# Example usage
units_consumed = 250
total_bill, status = calculate_electricity_bill(units_consumed)

print("Total Bill: ₹", total_bill)
print("Usage Status:", status)
```
