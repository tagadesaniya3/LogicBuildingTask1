```python
# Take input from user
sentence = input("enter sentence: ")

# Split sentence into words
words = sentence.split()

# Convert list to set to get unique words
unique_words = set(words)

# Print results
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

```python
# Employee dictionary
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

# Find employee with highest salary
highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

# Print result
print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

```python
# Given list
numbers = [45, 22, 89, 10, 66]

# Initialize max and min with first element
maximum = numbers[0]
minimum = numbers[0]

# Loop through the list
for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

# Print results
print("List:", numbers)
print("Max:", maximum)
print("Min:", minimum)
```

```python
# Given list of product prices
prices = [450, 1200, 899, 1500, 300]

# Initialize counter
count = 0

# Loop through prices
for price in prices:
    if price > 1000:
        count += 1

# Print result
print("Products above 1000:", count)
```

```python
# Given attendance list
attendance = ["P", "P", "A", "P", "P"]

# Count total days
total_days = len(attendance)

# Count number of presents
present_days = attendance.count("P")

# Calculate percentage
percentage = (present_days / total_days) * 100

# Print result
print("Attendance Percentage:", percentage)
```

```python
# Given list of phone numbers
phone_numbers = [9876543210, 9123456789, 9876543210]

# Convert list to set to remove duplicates
unique_numbers = set(phone_numbers)

# Print result
print("Unique phone numbers:", unique_numbers)
```

```python
# Given string
text = "pythonp"

# Create empty dictionary
frequency = {}

# Count each character
for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

# Print result
print(frequency)


```

```python
# Given list of marks
marks = [70, 75, 80, 65]

# Calculate total and count
total = sum(marks)
count = len(marks)

# Calculate average
average = total / count

# Print result
print("Average Marks:", average)
```
