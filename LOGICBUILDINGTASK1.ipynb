```python
username = "admin"
passward = "1234"

username1 = input("Enter username:")
passward2 = input("Enter passward:")
if username1 == username and passward2 == passward:
    print("Login successfully")
else:
    print("invalid Credentials")
```

```python
marks = [45, 78, 90, 33, 60]
Pass = 0
fail = 0
for m in marks:
    if m >= 50:
        Pass += 1
    else:
        fail += 1
print("Total Student:", len(marks))
print("Number of Students Passed:", Pass)
print("Number of Students Failed:", fail)
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = []

for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

print("origional name:", names)
print("Cleaned Names:", cleaned_names)

```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    print("Message:", message)
    print("Length:", length)

    if length > 10:
        print("Flag: Message is longer than 10 characters")

```

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

error_count = 0

for log in logs:
    if log == "ERROR":
        error_count += 1

print("Total ERROR entries:", error_count)
```
